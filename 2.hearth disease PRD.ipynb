# K FOLD CROSSVALIDATION¶

```python
# Import necessary libraries
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import numpy as np # Assuming x and y are numpy arrays or pandas objects

# Sample data (assuming x and y are defined elsewhere in your code)
# Example data for demonstration purposes:
# x = np.random.rand(100, 10)
# y = np.random.randint(0, 2, 100)

# Scale data
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Instantiate the model (create an object)
svc_model = SVC() # You can add parameters here, e.g., kernel='linear'
#### make svc_model = SVC() and knn_model = KNeighborsClassifier() of model befire using them in score = crossvalscore(knn_model etc)
# Use cross validation
# The 'svc_model' variable (an instance) is passed as the estimator
scores = cross_val_score(knn_model, x_scaled, y, cv=5, scoring='accuracy')

# Print the scores
#print(scores)
# Print the mean and standard deviation of the scores for a single metric
# print(f"Mean accuracy: {scores.mean():.2f} with a standard deviation of {scores.std():.2f}") TO GET STD OF DATA
print(scores.mean())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# feature scaling

```python
# in scaling u scale x test and x train cuz y is the one u need to predict
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
# making a dictionary for models
models = {
    "SVM ": SVC(probability = True),
    "Decision tree":DecisionTreeClassifier(),
    "Naive Byeas":GaussianNB(),
    "KNN":KNeighborsClassifier(),
    "LogisticRegression" : LogisticRegression(),

}
```

```python
# make a result tab so u can save results of model here
results=[] 
```

```python
from sklearn.metrics import accuracy_score, f1_score

for name, model in models.items(): # Use .items() for dictionary, or iterate list
    model.fit(x_train_scaled, y_train) # Fit the specific model
    y_pred = model.predict(x_test_scaled) # Predict on scaled test features (X_test)
    acc = accuracy_score(y_test, y_pred) # Compare predictions with true test labels (y_test)
    f1 = f1_score(y_test, y_pred, average='binary') # FIX: Add average='binary' for binary classification
    results.append({
        'Model': name,
        'Accuracy': round(acc, 4),
        'F1': round(f1, 4)
    })

```
